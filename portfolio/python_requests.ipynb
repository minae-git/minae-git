```python
def get_season_leaders_board(season_id_list, game_mode):
    
    ### packages required
    import requests
    import time
    
    ### local variable for requesting
    result_season = {}
    result_mode = {}
    prefix_url = "https://api.pubg.com/shards/pc-krjp/leaderboards/{}/{}"
    # url example : https://api.pubg.com/shards/pc-krjp/leaderboards/division.bro.official.pc-2018-02/squad-fpp
    header = {"accept": "application/vnd.api+json", 
           "Authorization": "Bearer eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJqdGkiOiI4NzQ3OGVjMC1kYWMxLTAxMzgtMzE0YS02MzY4NzhjMmExNzkiLCJpc3MiOiJnYW1lbG9ja2VyIiwiaWF0IjoxNjAwMzEyMzQyLCJwdWIiOiJibHVlaG9sZSIsInRpdGxlIjoicHViZyIsImFwcCI6Im1pbmFlMjMyNCJ9.TNu-n4s9e_gIH0pNC5F6DtibY2ETNah_0QScxQGhzEs"}
    
    
    for season_id in season_id_list[:8]: # for pc-2018
        for mode in game_mode:
            url = prefix_url.format(season_id, mode)
            
            # For checking the process
            print(url)
            
            #   
            r = requests.get(url, headers=header)
            # only 10 requests per minute are allowed
            time.sleep(6)
            
            # store the JSON reuslts specified by game mode
            result_mode[mode] = r.json()
            
        # Store the results of certain season
        result_season[season_id] = result_mode
        
    return result_season            
```

```python
leaders_board_results = get_season_leaders_board(season_id, game_mode)
```
