## Ch1. Importing Data From Flat Files

```python
import pandas as pd
```

```python
tax_2016 = pd.read_csv('../data/vt_tax_data_2016.csv')
```

```python
tax_2016.head()
```

```python
# usecols

cols = ['zipcode', 'agi_stub', 'mars1', 'MARS2', 'N2']
tax_2016_usecols = pd.read_csv('../data/vt_tax_data_2016.csv', usecols=cols)
tax_2016_usecols.head()
```

```python
# skiprows, nrows

print(tax_2016.shape)
tax_2016_rows = pd.read_csv('../data/vt_tax_data_2016.csv', skiprows=100, nrows=500)
print(tax_2016_rows.shape)
```

```python
# dtype <- dictionary, data type should be passed in quotation
print(tax_2016_usecols.info()) # zipcode -> int

print("\n", "*"*40)
tax_2016_usecols_dtype = pd.read_csv('../data/vt_tax_data_2016.csv', usecols=cols, dtype={'zipcode':str})
print(tax_2016_usecols_dtype.dtypes)
```

```python
# na_values

print(tax_2016_usecols.head(10)) # zipcode = 0 -> bad data, treated as NA
print('\n', "*"*40)

tax_2016_na_values = pd.read_csv('../data/vt_tax_data_2016.csv', na_values={'zipcode':0}, usecols=cols)
# zipcode 중에서 0인 애들은 모두 na로 변경
print(tax_2016_na_values.isna().sum())
```

```python
# import error handling -- error_bad_lines, warn_bad_lines

# 문제가 있는 row를 제외한 data로 DF 생성, bad line있을 경우 print
tax_2016_error = pd.read_csv('../data/vt_tax_data_2016.csv', error_bad_lines=False, warn_bad_lines=True)
```
