##### 2019년 의 영화 평점 데이터를 가져옵니다.

```python
import time

date = pd.date_range('20190101', periods=31, freq='D')

movie_date = []
movie_name = []
movie_point = []

for today in date:
    html = "http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=
                                   urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
    time.sleep(1)
```

```python
# 수집된 데이터 건수 확인
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date':movie_date, 
                      'name':movie_name, 
                      'point':movie_point})
movie.head()
```

```python
# Visualization (영화 랭킹)

# 150일 이상 랭킹에 포함된 영화 중 상위 5개 영화의 평균 평점과 평점 산포도

rank_in_days = movie['name'].value_counts()
mv_over_200days = rank_in_days[rank_in_days.values >= 150].index[:5]
print(mv_over_200days)

movie1 = movie.set_index(['name', 'date'])
movie1 = movie1.loc[mv_over_200days].reset_index()
```

```python
plt.figure(figsize=(15,3))

for i in range(5):
    # 영화별 평점
    name_filter = movie1['name']==mv_over_200days[i]
    x_data = np.arange(movie1[name_filter].shape[0])
    y_data = movie1[name_filter]['point']
    
    # 평균 평점
    y_data_mean = np.mean(y_data)
    
    # 영화별 subplot
    plt.subplot(1,5, i+1)
    plt.title(mv_over_200days[i])
    plt.plot(x_data, y_data)
    plt.axhline(y=y_data_mean, color='r')
```
